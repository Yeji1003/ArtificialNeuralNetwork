### 1.2 Stateful vs stateless approach

You will often see two different approaches when you look for plotting examples on the internet. In these examples either the stateful (state-based) or the stateless (object-oriented) approach is used. The differences between them may not be clear at first sight, but they will become clearer with the following plotting example.


**The stateful approach:**

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot([0,1], [1,0])
```

**The stateless approach:**

```python 
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([0,1], [1,0])
```

Both of the approaches carry out the same steps:

- Create a figure object (with a plotting area).
- Create a line plot into the plotting area.

The only difference lies within the object-oriented form. In the stateless approach we do the operations above rather explicitly and take the object references into variables (fig, ax). Calling methods of the axes object (ax) we can directly modify the underlying objects. The references are not lost in the stateful approach, either, but they are stored somewhere behind the scenes. There is only ever one figure or axes that you’re manipulating at a given time, and you don’t need to explicitly refer to it.

The stateful approach is really good for creating quick plots. But when it comes to more complex plots it will get messy very quick. Thus we will concentrate mostly on using the stateless approach in this course.

### 1.3 How to create single plots

The most common way to create a figure with a single axes (plotting area) is with [```plt.subplots()```](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html):  

```python
fig, ax = plt.subplots()
```

Notice that if we don't pass any arguments to `plt.subplots()`, so the default call is executed: `plt.subplots(nrows=1, ncols=1)`. Consequently, ax is created as a single AxesSubplot object, that we can modify with different methods. We will show some of these methods in the following example:

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# define the number of x-coordinates
x = np.arange(4)

# create some data
money = [100, 400, 600, 1500]
owner = ['Bill', 'Fred', 'Mary', 'Sue']

# create a figure object with a single axes object
fig, ax = plt.subplots()

# create a bar plot of the data
ax.bar(x, money)

# set the ticks on the x-axis and label them
ax.set_xticks(x)
ax.set_xticklabels(owner)

# set the ticks on the y-axis
ax.set_yticks(np.arange(0, 1600, 250))

# add some general descriptions
ax.set_title('Savings', fontsize=12)
ax.set_ylabel('[$]')

plt.show()
````

### 1.4 How to create multiple plots

For creating a figure with multiple plots we first have to take another look on [```plt.subplots()```](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html). For single plots we simply used this command with the default values ```nrows=1``` and ```ncols=1```, which created a 1x1 subplot grid for a single plot. Which means that by specifying the number of rows and columns we can create a subplot grid for several axes objects. If we create e.g. a subplot of 2 rows and 2 columns we get 4 axes (plotting areas).

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# create a figure object with 4 axes objects and a predefined size 
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(7,5))
````

Each axes object is accessible by its index and can be modified separately.

````python
for i in range(2):
    for j in range(2):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')

fig
````

Using the indices of each axes object we can easily fill in the plots in the plotting areas.

````python
# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# create a figure object with 4 axes objects and a predefined size 
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7,5))

# create the plots for each axes object
axes[0,0].plot(x, y)
axes[0,1].plot(x, y**2, color='orange')
axes[1,0].plot(x, -y, color='green')
axes[1,1].plot(x, -y**2, color='red')

plt.show()
````

## 2. The Advanced: Plotting with Seaborn

Now that the basics of matplotlib are clear we will concentrate on creating basic plots using real data from [github](https://github.com/mwaskom/seaborn-data). Therefore we will go through some seaborn plots that you can use for data visualization. Note that most of the plots we will use in seaborn can also be created with matplotlib. But seaborn is with its dataset-oriented plotting functions that internally perform the necessary semantic mappings way easier to handle than matplotlib, producing informative complex plots with only a few lines of code.

### 2.1 Load package imports and dataset

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
````

````python
# use seaborn to load the dataset 'tips' as a pandas dataframe
tips = sns.load_dataset('tips')
````

````python
# take a look at the head of the dataframe
tips.head()
````

````python
# take a look at the tail of the dataframe
tips.tail()
````

### 2.2 Barplots

You can use Seaborn's [```barplot()```](https://seaborn.pydata.org/generated/seaborn.barplot.html) function to plot simple barplots. <br>
Some of the more important parameters that you should have in mind are:

- **capsize:** Width of the “caps” on error bars.
- **hue:** Grouping variable that will produce elements with different colors. 
- **order:** Order to plot the categorical levels in.
- **ax:** If provided, plot on this axis.

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# create barplots
sns.barplot(data=tips, x='day', y='total_bill', ax=axes[0], capsize=0.2)
sns.barplot(data=tips, x='day', y='total_bill', ax=axes[1], capsize=0.2, 
            hue='sex', order=['Sat','Sun','Thur','Fri'])

# add some general descriptions
axes[0].set(title='Barplot', xlabel='', ylabel='total bill')
axes[1].set(title='Barplot', xlabel='', ylabel='total bill')

plt.show()
````

### 2.3 Boxplots

You can use Seaborn's [```boxplot()```](https://seaborn.pydata.org/generated/seaborn.boxplot.html) function to show distributions with respect to categories. <br>
Some of the more important parameters that you should have in mind are:

- **hue:** Grouping variable that will produce elements with different colors.
- **order:** Order to plot the categorical levels in.
- **ax:** If provided, plot on this axis.

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# create boxplots
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0])
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[1], 
            hue='sex', order=['Sat','Sun','Thur','Fri'])

# add some general descriptions
axes[0].set(title='Boxplot', xlabel='', ylabel='total bill')
axes[1].set(title='Boxplot', xlabel='', ylabel='total bill')

plt.show()
````

### 2.4 Distplots

You can use Seaborn's [```distplot()```](https://seaborn.pydata.org/generated/seaborn.distplot.html) function to plot histograms and joint distributions of variables. <br>
Some of the more important parameters you should have in mind are:

- **bins:** Specification of hist bins. 
- **hist (bool):** Whether to plot a (normed) histogram.
- **kde (bool):** Whether to plot a gaussian kernel density estimate.
- **rug:** Whether to draw a rugplot on the support axis.
- **axlabel:** Name for the support axis label. If False, do not set a label.
- **ax:** If provided, plot on this axis.

````python
#create figure and axes objects of a predefined size with a shared x-axis
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, figsize=(10,8))

# create distplots
sns.distplot(tips['total_bill'], ax=axes[0,0], axlabel=False)
sns.distplot(tips['total_bill'], ax=axes[0,1], axlabel=False, kde=False)
sns.distplot(tips['total_bill'], ax=axes[1,0], axlabel='total bill', hist=False)
sns.distplot(tips['total_bill'], ax=axes[1,1], axlabel='total bill', hist=False, rug=True)

# add some general descriptions
axes[0,0].set(title='Histogram with KDE')
axes[0,1].set(title='Histogram without KDE')
axes[1,0].set(title='KDE without Histogram')
axes[1,1].set(title='KDE with Rugplot')

plt.show()
````

### 2.5 Scatterplots

You can use Seaborn's [```scatterplot()```](https://seaborn.pydata.org/generated/seaborn.scatterplot.html) function to plot relationships between x and y showing different subsets of the data using hue, style and size. Some of the more important parameters that you should have in mind are:

- **hue:** Grouping variable that will produce elements with different colors.
- **style:** Grouping variable that will produce points with different markers.
- **size:** Grouping variable that will produce elements with different sizes. 
- **sizes:** E.g. a list of size values mapping levels of the size variable to sizes.
- **ax:** If provided, plot on this axis

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# create scatterplots
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[0])
# depict days with a different color
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[1], hue='day')

# add some general descriptions
axes[0].set(title='Scatterplot', xlabel='total bill')
axes[1].set(title='Scatterplot', xlabel='total bill')

plt.show()
````

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# depict the dining times with a different marker
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[0], style='time')
# depict the dining times with a different marker and the days with a different color
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[1], style='time', hue='day')

# add some general descriptions
axes[0].set(title='Scatterplot', xlabel='total bill')
axes[1].set(title='Scatterplot', xlabel='total bill')

plt.show()
````

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# depict the guest count regarding their sizes
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[0], size='size')
# depict the guest count with predefined sizes and different colors
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[1], size='size', 
                sizes=(20,200), hue='size')

# add some general descriptions
axes[0].set(title='Scatterplot', xlabel='total bill')
axes[1].set(title='Scatterplot', xlabel='total bill')

plt.show()
````

### 2.6 Violinplots

You can use Seaborn's [```violinplot()```](https://seaborn.pydata.org/generated/seaborn.violinplot.html) function to draw a combination of boxplot and kernel density estimate. Some of the more important parameters that you should have in mind are:

- **hue:** Grouping variable that will produce elements with different colors.
- **order:** Order to plot the categorical levels in.
- **split:** When using hue nesting with a variable that takes two levels, setting split to True will draw half of a violin for each level. This can make it easier to directly compare the distributions.
- **inner:** Representation of the datapoints in the violin interior. If 'box', draw a miniature boxplot. If 'quartiles', draw the quartiles of the distribution. If 'point' or 'stick', show each underlying datapoint. 
- **ax:** If provided, plot on this axis

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# create violinplots
sns.violinplot(data=tips, x='day', y='total_bill', ax=axes[0])
sns.violinplot(data=tips, x='day', y='total_bill', ax=axes[1], 
               order=['Sat','Sun','Thur','Fri'], inner='quartile')

# add some general descriptions
axes[0].set(title='Violinplot', xlabel='', ylabel='total bill')
axes[1].set(title='Violinplot', xlabel='', ylabel='total bill')

plt.show()
````

````python
#create figure and axes objects of a predefined size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# create violinplots
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', ax=axes[0])
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', ax=axes[1], split=True)

# add some general descriptions
axes[0].set(title='Violinplot', xlabel='', ylabel='total bill')
axes[1].set(title='Violinplot', xlabel='', ylabel='total bill')

plt.show()
````
